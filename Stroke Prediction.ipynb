# Data Understanding

## This dataset is provided from https://www.kaggle.com/fedesoriano/stroke-prediction-dataset. It is a free, and publiclly available dataset and does not contain any personal information from patients. 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier 
from sklearn import metrics, model_selection
from sklearn.ensemble import BaggingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
import textwrap
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
import IPython.display
from IPython.display import display, clear_output
import plotly.graph_objects as go
```

### This is a brief glimpse into the dataset, which contains 5110 rows and 12 columns of data

```python
df = pd.read_csv("C:/Users/Tara/Downloads/healthcare-dataset-stroke-data.csv")

# Print portion of the data
display(df)
```

```python
# Number of rows and columns
df.shape
```

```python
# View the column names, number of entries in each column, and data types
df.info()
```

```python
# Calculate the number of null entires within the bmi column
sum(df.bmi.isnull())
```

```python
# Use mean imputation to resolve missing data
mean = df.bmi.mean()
df.loc[df['bmi'].isnull(),'bmi'] = mean
```

```python
# Verify that there is no missing data in the new dataframe
df.info()
```

```python
# Since the columns for gender, ever_married, work_type, residence_type, and smoking_status
# are categorical rather than numerical, label encoding will assign codes to each distinct entry.
le=preprocessing.LabelEncoder()
```

```python
df['gender'] = le.fit_transform(df['gender'])
df['smoking_status'] = le.fit(df['smoking_status'])
df['work_type'] = le.fit_transform(df['work_type'])
df['Residence_type'] = le.fit_transform(df['Residence_type'])
df['smoking_status'] = le.fit_transform(df['smoking_status'])
df['ever_married'] = le.fit_transform(df['ever_married'])
```

```python
# Verify that there are no object data types within the dataframe
df.info()
```

# Exploratory Data Analysis

## The dataset had 4.9% of patients suffer from stroke.

```python
# Set the target and feature variables
# Seperate the stroke column, as this will be the target variable
X = df.drop(['stroke'], axis=1)
y = df['stroke']
```

```python
# Visualize the data  
def get_fig():
    plt.figure(figsize=(7,7))
    plt.pie(df['stroke'].value_counts(sort = True),
        explode = (0.05, 0), 
        labels = df['stroke'].value_counts(sort = True).index,
        colors = ["green","orange"], 
        autopct = '%1.1f%%')
    plt.title('Percentage of Participants with Stroke')
    plt.show
get_fig()

# This shows us that 4.9% of respondents had a stroke.
```

## Age, heart disease, avgerage glucose level, hypertension, and ever married have the highest correlation with stroke, with age being the variable with the highest correlation. Also, id, gender, work type, and residence type have the least correlation. 

```python
# Now lets visualize how the feature variables correlate with the target variable.
X.corrwith(y).plot(kind='bar', grid=True, figsize=(25, 15), title="Correlation with Target Variable (Stroke)")

# Age, heart_disease, avg_glucose_level, hypertension, and ever_married have the highest correlation with stroke, with age being the 
# variable with the highest correlation with stroke. Also, id, gender, work_tye, and residence_type have the least correlation. 
```

## Each variable's correlation with the stroke can be visualized with a darker blue box representing higher correlation. Again, age is shown to have the highest correlation with stroke.

```python
# Another way to visualize the correlation is through a correlation matrix 
plt.figure(figsize=(30,20))
sns.heatmap(df.corr(), cmap = 'Blues',annot=True)
```

```python
# We split the data between 80% training and 20% testing for model validation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# View the dimensions of the test and training datasets
print("Dimensions of the X_train dataset: ", X_train.shape)
print("Dimensions of the y_train dataset: ", y_train.shape)
print("Dimensions of the X_test dataset: ", X_test.shape)
print("Dimensions of the y_test dataset: ", y_test.shape)
```

```python
# Now we can scale the data to remove bias from the model
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Modeling

## I will also have a dropdown menu to choose the model and the results will be displayed for the chosen model.

### The K-nearest neighbors model has an accuracy of 93.8%. 

```python
# First we will try K-nearest neighbors for target varaible prediction
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train, y_train)

y_pred_knn = knn_clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred_knn)

sns.heatmap(cm, center = True, fmt='', cmap='Blues', annot=True)
plt.title('Confusion Matrix')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
print("K-NN Accuracy Score:\t", accuracy_score(y_test, y_pred_knn))

# The k-nearest neighbors model has a 93.8% accuracy and with 1 false positive and 62 false negatives.
```

```python
from sklearn.svm import SVC

# We can also apply a support vector machine (SVM) model 
SVM = SVC(random_state = 0, kernel = 'linear')
SVM.fit(X_train, y_train)
y_pred_SVM = SVM.predict(X_test)

cm = confusion_matrix(y_test, y_pred_SVM)

sns.heatmap(cm, center = True, fmt='', cmap='Blues', annot=True)
plt.title('Confusion Matrix')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
print("SVM Accuracy Score:\t", accuracy_score(y_test, y_pred_SVM))
```

### The SVM model has an accuracy of 93.9%.

```python
# Decision Trees Classifier

dt = DecisionTreeClassifier(random_state=42)
dt = dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

cm = confusion_matrix(y_test, y_pred_dt)

sns.heatmap(cm, center = True, fmt='', cmap='Blues', annot=True)
plt.title('Confusion Matrix')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
print("Decision Tree Accuracy Score:\t", accuracy_score(y_test, y_pred_dt))

# Decision tree classifier has a 91.4% accuracy
```

### The Decision Tree classifier has an accuracy of 91.4%

```python
from sklearn.linear_model import LogisticRegression 

log_reg = LogisticRegression(solver = 'liblinear')
log_reg.fit(X_train, y_train)

y_pred_log = log_reg.predict(X_test)

cm = confusion_matrix(y_test, y_pred_log)

sns.heatmap(cm, center = True, fmt='', cmap='Blues', annot=True)
plt.title('Confusion Matrix')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
print("Logistic Regression Accuracy Score:\t", accuracy_score(y_test, y_pred_log))

# Logistic Regression has a 91.4% accuracy
```

### Logistic Regression has an accuracy of 93.9%.

```python
# Random Forest Classifier 

rf = RandomForestClassifier(n_estimators = 25)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

cm = confusion_matrix(y_test, y_pred_rf)

sns.heatmap(cm, center = True, fmt='', cmap='Blues', annot=True)
plt.title('Confusion Matrix')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
print("Random Forest Accuracy Score:\t", accuracy_score(y_test, y_pred_rf))

# Random forest classifier has a 93.6% accuracy
```

### Random Forest has an accuracy of 93.6%

```python
#Gaussian naive bayes model
nb = GaussianNB()
nb.fit = nb.fit(X_train,y_train)

y_pred_nb = nb.predict(X_test)

cm = confusion_matrix(y_test, y_pred_nb)

sns.heatmap(cm, center = True, fmt='', cmap='Blues', annot=True)
plt.title('Confusion Matrix')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
print("Naive Bayes Accuracy Score:\t", accuracy_score(y_test, y_pred_nb))

# Gaussian Naive Bayes model has a 87.2% accuracy
```

### Gaussian Naive Bayes has an accuracy of 87.1%.

```python
# Bagging Classifier
from sklearn import ensemble

model = ensemble.BaggingClassifier()
model.fit(X_train, y_train)
y_pred_bag = model.predict(X_test) 

sns.heatmap(cm, center = True, fmt='', cmap='Blues', annot=True)
plt.title('Confusion Matrix')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
print("Bagging Accuracy Score:\t", accuracy_score(y_test, y_pred_bag))
```

### Bagging Classifier has an accuracy of 93.7%.

```python
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier(learning_rate=0.1, n_estimators=50, random_state=10)
clf.fit(X_train, y_train)
y_pred_xg = clf.predict(X_test)

sns.heatmap(cm, center = True, fmt='', cmap='Blues', annot=True)
plt.title('Confusion Matrix')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
print ("XGBoost Accuracy Score:\t",accuracy_score(y_test, y_pred_xg))
```

### XGBoost has an accuracy of 94.0%

# Conclusion
